Travel time is the dependent variable and kilometers travelled and number of deliveries are the independent variables. 
Dependent = Response, Independent = Predictor

New considerations when moving from simple to multiple regression:
1. Adding more independent variables to a multiple regression does not mean that the regression is better or that it will give better predictions, in fact it can be worse. This is called OVERFITTING
2. The addition of more independent variables creates more relationships between them. So not only do we need to consider the independent variables and the dependent variables, we also need to consider the relationship between the independent variables. This is called MULTICOLINEARITY.
3. The ideal is for all independent variables to be correlated with the dependent variable and NOT with each other.
4. Because of overfitting and mutlicolinearity, there is a lot of prep work before we start multiple regression.
   
   - Scatterplots
   - Correlations
   - Simple regressions between each of the independent variables and the dependent variables to see how they are related
  
Important to consider which independent variables make the cut. Some independent or sets of independent variables are better at predicting the dependent variables than others. Some contribute nothing.

Mutliple regression model: $\beta_0 + \beta_1x_1+\beta_2x_2+\dots+\beta_px_p + \epsilon$

$y = 27 + 9x_1 + 12 x_2$

- x_1 = capital investment (€1000s)
- x_2 = marketing expenses (€1000s)
- y = predicted sales (€1000s)

Each coefficient (9, 12) is interpreted as the estimate in y corresponding to a 1 unit change in variable when all other variables are held fixed.

€9000 is an estimate of the expected increase in sales y, corresponding to a €1000 increase in capital investment $x_1$ when marketing expenditure $x_2$ is held fixed

Nice exam question. Conducting multiple regression requires the following steps:
1. Generate a list of independent and dependent variables
2. Collect data on the variables
3. Check for outliers
4. Check the relationships between each independent variable and the dependent variable using scatterplots and correlations
5. Check the relationships between the independent variables using scatterplots and correlations
6. Conduct simple linear regressions for each IV/DV pair
7. Use the non-redundant independent variables in the analysis to find the best fitting model
8. Use the best fitting model to make predictions about the dependent variable

How to we choose the best model?
1. $R^2 adjusted$
2. The standard error
3. The AIC and BIC to be as low as possible
4. Consider the p-value for all of the coefficients in the model. If they are not all significant this indicates a problem
5. Consider the model. Does it make sense? Often is the coefficients are not significant the model will not make sense when you consider the problem. When the coefficients are not significant they are basically random.
6. If the condition number is very high that is a good indicator of problems
7. All else being equal the simplest model is the best

<b> Total sum of squares </b>
- SST is the sum of the squared distances from the mean model (a horizontal line having slope 0 at the mean of the prices) and each observation
- Once the data is locked in w.r.t the total number of observations the total sum of squares is set
- SST is a finite sum that does not change regardless of how many features we add into the regression model: SST is not feature dependent, it is row dependent
- Feature variables partition (allocate) the total sum of squares to the model and the remaining sum of squares in error (residuals)
- As features are added or removed the partitioning of the SSE and the SSR will almost certainly change
- How the sum of squares is allocated to each feature with the SSR will change as the model changes
- If no features have significant explanatory power on y, the model will have zero features, and the best prediction is always $\bar{y}$, the horizontal line with slope 0

SST is split into SSR(Regression model sum of squares) and SSE(residual error sum of squares).
<b>$R^2 = SSR/SST</b> it describes the proportion of the sum of squares that we can put in the model. The goal is to:
- Maximise SSR as a proportion of SST
- Minimize SSE as a proportion of SST

- SSR is then split among the features of the model in terms of relative explanatory power
- Not only is SSR likely to change but the allocation of SSR to each feature in the model changes are features are added and removed
- The SST is fixed regardless of the number of features in the model

<b> Overfitting and SSR </b>
- As more features are added to the model the SSR will never decrease; SSR can can only stay the same of get better
- As additional features are added, the mode starts to shape itself to the data that it is trained on, keeps optimising
- The model tries to squeese out as much SSE as possible whether the feature adds predictive power or not
- Complex regressions could suffer on new data because the nooks and crannies of the data are absent
- There are techniques to avoid overfitting:
- Model building procedures (Forward selection, Backward elimination, Stepwise regression, Best subsets regression) along with AIC, BIC, and Mallows C_p.
- Regularisation penalises complex models
- Cross validation
- PCA Principal Component Analysis


<b> Multiple Regression </b>
There are three common techniques for building iterative regression models:
1. Forward Selection
2. Backward Elimination
3. Stepwise Regression

There is fourth common technique that is not iterative. It is called best subsets regression. Best subsets examines all possible combinations of feature variables; brute force method that can computationally expensive.
- Best subsets will examine all of the options, possibly many
- The analyst can specify the maximum number of features
- The analyst can also request the best two or three models for each number of feature variables

<b> Forward Selection </b>
- Features variables are examined on at a time
- The analyst sets a stopping rule (p-value) that acts as a hurdle the variable has to overcome to get into the model
- The added variable must reduce the error (SSE) significantly
- The feature variable that reduces the model error (SSE) the most is chosen as long as it passes the stopping rule
- Once a variable is added to the model it is never removed
- The process repeats itself until all variables are in the model OR no variable passes the stopping rule

<b> Flaws with forward selection </b>
- The ability of variables to reduce error can change as other variables enter the model
- As feature variables are added they begin to overlap and interact in how they explain variance
- Since a feature variable stays in the model, once it is selected, forward selection is not flexible
- It is possible for a variable that entered the model early to fail the stopping rule later as other variables are added, but it is stuck in the model
- Example: $V1$ enters first. Then $V2$. Then $V3$. Let's say that $V2$ and $V3$ combined reduce the error better than $V1$ alone. This could lead to $V1$ failing the stopping rule, but it is stuck in the model
- Squared semipartial correlations can tease out these relationships
- It can miss supressor and/or complimentary relationships:
  - Suppressor variables (also called enhancer variables) are strange. They correlate with parts of other variables but not the target. They eat up (suppress) some of the non-relvant error in those variables making the model stronger.
  - Complimentary variables are negatively correlated with each other but together they explain the target variable well
